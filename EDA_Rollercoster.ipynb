```python
#Renaming columns
df=df.rename(columns={'coaster_name':'Coaster_name','year_introduced':'Year_introduced'})
df.head()
```

```python
#Barplot
ax = df['Year_introduced'].value_counts().head(10).plot(kind='bar',title = 'Top 10 Years of coaster introduction')
ax.set_xlabel('Year')
ax.set_ylabel('Number of Coasters')
```

```python
#Histogram
ax1 = df['speed_mph'].plot(kind='hist',title = 'Speed of Coasters')
ax1.set_xlabel('Speed')
ax1.set_ylabel('Number of Coasters')
```

```python
ax1 = df['speed_mph'].plot(kind='kde',title = 'Speed of Coasters')
ax1.set_xlabel('Speed')
ax1.set_ylabel('Number of Coasters')
```

```python
#Scatter plot matplotlib
df.plot(kind='scatter',x='speed_mph',y='height_ft',title='Speed vs Height')
```

```python
#Seaborn scatterplot
sns.scatterplot(x='speed_mph',y='height_ft',data=df)
```

```python
#Pairplot with multiple columns
sns.pairplot(df,vars=['Year_introduced','speed_mph','height_ft','Inversions','G-force'])
plt.show()
```

```python
df_corr = df[['Year_introduced','speed_mph','height_ft','Inversions']].dropna().corr()
df_corr
```

```python
#Heatmaps
sns.heatmap(df_corr,annot=True)
```
